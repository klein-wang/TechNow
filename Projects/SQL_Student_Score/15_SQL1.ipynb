### SQL script
There are two tables in the database: 
1. __`student_score_feb`__ with the format as 
    ```
        {
        student_id  BIGINT(255)     student index; related to database.students
        score       DECIMAL(4,2)    score the student achieved in the fed exam 
        subject     VARCHAR(20)     subject {'Math', 'Chinese', 'English'}
        }
    ```
2. __`students`__ with the format as :
    ```
        {
        student_id   BIGINT(255)   student index; related to database.student_score_feb
        student_name VARCHAR(20)   name of the student
        parent_tel   VARCHAR(20)   tel-num of one of the student's parents
        }
    ```

#### Q1
please find out the top 10% student ids in Math(student_score_feb.subject = "Math")

````python
import sqlite3 as sql
import pandas as pd
````

````python
con = sql.connect('data.db')
cur = con.cursor()
````

````python
# please write you code here
query = """
SELECT * 
from student_score_feb
where subject = 'Math'
order by score desc
limit round((Select 0.1*count(*) from student_score_feb where subject = "Math"))
"""
pd.read_sql(query, con = con)
````

#### Q2
please find out those who failed in the feb exam (student_score_feb.score < 60);  
list the student names, their parents' phone numbers and the number of failed subjects

````python
# please write you code here
query = """
SELECT b.student_name, 
       b.parent_tel,
       count(a.subject) as No_Fails
FROM student_score_feb a 
    JOIN students b ON a.student_id = b.student_id
WHERE a.score < 60
GROUP BY student_name,parent_tel
"""

pd.read_sql(query, con = con)
````
